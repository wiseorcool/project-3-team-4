```python
#This entire code is to create a new table for genre id and description of the genre. 
# Genre is subset from the movies tables

# Extract the 'genres' column
data_genre_raw = df_movie_metadata[['genres']].copy()


# Function to convert string representation of list to actual list
def str_to_list(s):
    return ast.literal_eval(s)


# Apply the function to the 'genres' column
data_genre_raw['genres'] = data_genre_raw['genres'].apply(str_to_list)


# Flatten the list of lists into a single list of dictionaries
flattened_genres = [genre for sublist in data_genre_raw['genres'] for genre in sublist]
# Create a DataFrame from the flattened list
df_genres = pd.DataFrame(flattened_genres)

# Remove duplicates
df_unique_genres = df_genres.drop_duplicates().reset_index(drop=True)

# Inspect the resulting DataFrame
print("DataFrame with genre_Id and Name of genre_Id:")
print(df_unique_genres)
```

```python
# This part of the code is create a relationship table between movies table and genre_id
# Genre is subset from the movies tables

movieId_genreId_df = df_movie_metadata[['genres', 'id']]
#function to convert string to list
def str_to_list(s):
    return ast.literal_eval(s)

# Apply the function to the 'genres' column
movieId_genreId_df['genres'] = movieId_genreId_df['genres'].apply(str_to_list)

# Create a new column with genre_ids only
movieId_genreId_df['genre_ids'] = movieId_genreId_df['genres'].apply(lambda genres: [genre['id'] for genre in genres])

# Use explode to split lists into separate rows with one genre_id in each row
movieId_genreId__final_df = movieId_genreId_df.explode('genre_ids')

# # Rename columns for clarity
movieId_genreId__final_df.rename(columns={'genre_ids': 'genre_id'}, inplace=True)
#Drop the column genres as from the data frame
movieId_genreId__final_df.drop(columns=['genres'],inplace=True)

# movieId_genreId__final_df = movieId_genreId__final_df.drop_duplicates().reset_index(drop=True)

# # Display the resulting DataFrame
print(movieId_genreId__final_df.head(10))
```

```python
#this entire code is to create a new table for production id and name of production company 
# production company is subset of the movies tables

# Extract the 'genres' column
data_production_company_raw = df_movie_metadata[['production_companies']].copy()
data_production_company_cleaned=data_production_company_raw.dropna()

# Function to convert string representation of list to actual list

def str_to_list(s):
    try:
        # Use ast.literal_eval to safely evaluate the string as a Python expression
        return ast.literal_eval(s)
    except (ValueError, SyntaxError):
        # If evaluation fails, return an empty list or appropriate value
        return []

# Apply the function to the 'genres' column
data_production_company_cleaned['production_companies'] = data_production_company_cleaned['production_companies'].apply(str_to_list)
data_production_company_cleaned = data_production_company_cleaned[data_production_company_cleaned['production_companies'].apply(lambda x: isinstance(x, list))]

# Flatten the list of lists into a single list of dictionaries
flattened_production = [product for sublist in data_production_company_cleaned['production_companies'] for product in sublist]
# Create a DataFrame from the flattened list
df_production = pd.DataFrame(flattened_production)

# # Remove duplicates
df_unique_production= df_production.drop_duplicates().reset_index(drop=True)

# # Inspect the resulting DataFrame
columns=['id','name']
df_unique_production=df_unique_production.reindex(columns=columns)
print("DataFrame with ID and Name:")
print(df_unique_production.head(2))

```

```python
# This part of the code is create a relationship table between movies table and production company id
# production company is subset from the movies tables

import pandas as pd
import ast

df_new = df_movie_metadata[['production_companies', 'id']].copy()

# Function to extract production company IDs and create a relationship table
def extract_relationships(df):
    relationships = []
    
    for index, row in df.iterrows():
        if row['id'] == 'id':  # Skip the header row
            continue
        production_companies = row['production_companies']
        if isinstance(production_companies, str):
            try:
                production_companies = ast.literal_eval(production_companies)
                if isinstance(production_companies, list):
                    for company in production_companies:
                        if isinstance(company, dict) and 'id' in company:
                            relationships.append({'movie_id': row['id'], 'company_id': company['id']})
            except (ValueError, SyntaxError):
                # Handle rows where production_companies cannot be parsed
                continue

    return pd.DataFrame(relationships)

# Create the relationship table
relationship_df = extract_relationships(df_new)

# Display the first few rows of the relationship table
print(relationship_df.head(2))
print(relationship_df.count())

# # Optionally, save the relationship table to a CSV file
# relationship_df.to_csv('relationship_table.csv', index=False)
```
